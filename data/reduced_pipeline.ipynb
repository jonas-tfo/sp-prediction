```python
import pandas as pd
```

```python
p6_dataframe = pd.read_csv("raw/complete_set.csv").drop(columns=["partition"])
p6_dataframe
```

```python
### Filter Data
# filter everything len(label) < 70
filtered_df = p6_dataframe[p6_dataframe["length"] == 70]
filtered_df
```

```python
# remove duplicates
filtered_df = filtered_df.drop_duplicates(keep="first").reset_index(drop=True)
filtered_df
```

```python
# remove entries with differing annotations
filtered_df = filtered_df.drop_duplicates(subset=["sequence"], keep=False).reset_index(drop=True)
filtered_df
```

```python
with open("input.fasta", "w") as f:
    for i, row in filtered_df.iterrows():
        f.write(f">{row['uniprot_id']}\n{row['sequence']}\n")
```

```python
%%bash
## external clustering using mmseqs
mmseqs createdb input.fasta inputDB
mmseqs cluster inputDB clusterRes tmp --min-seq-id 0.3
mmseqs createsubdb clusterRes inputDB repDB
mmseqs convert2fasta repDB reduced_30.fasta
```

```python
representatives = set()
with open("reduced_30.fasta", "r") as f:
    for line in f:
        if line.startswith(">"):
            representatives.add(line[1:].strip())

reduced_df = filtered_df[filtered_df["uniprot_id"].isin(representatives)].copy()
reduced_df
```

```python
## amino acid validation
valid_aa = set("ACDEFGHIKLMNPQRSTVWY")

for i, row in reduced_df.iterrows():
    if set(row["sequence"]) - valid_aa:
        print(f"{i}; {row['uniprot_id']}")
```

```python
## binary split for sp / no_sp classification
reduced_df.loc[:, "has_sp"] = reduced_df["type"] != "NO_SP"
reduced_df
```

```python
# train&val 80%, test 20%
from sklearn.model_selection import train_test_split

train_val_df, test_df = train_test_split(reduced_df, test_size=0.2, stratify=reduced_df["type"], random_state=69)
```

```python
train_df, val_df = train_test_split(train_val_df, test_size=0.2, stratify=train_val_df["type"], random_state=69)
```

```python
train_df
```

```python
import pandas as pd

counts = train_df["type"].value_counts()
max_count = counts.max()

dfs_to_concat = []

for label, count in counts.items():
    # Filter by the specific label
    train_subset = train_df[train_df["type"] == label]
    
    # Oversample to match max_coun replace=True so we can sample more rows than exist
    oversampled_subset = train_subset.sample(n=max_count, replace=True, random_state=69)
    
    dfs_to_concat.append(oversampled_subset)

# Concatenate at the end
train_df = pd.concat(dfs_to_concat, ignore_index=True)

# Shuffle the final dataframe so classes aren't ordered blocks
train_df = train_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
from sklearn.model_selection import StratifiedKFold

# Setup 3-fold cross validation
n_folds = 3
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

# Create fold indices
fold_splits = list(skf.split(train_val_df, train_val_df['type']))

print(f"Created {n_folds}-Fold Cross Validation")
print(f"CV dataset size: {len(train_val_df):,} sequences")
print(f"Test set size: {len(test_df):,} sequences (held out)\n")

# Show fold statistics
for fold_idx, (train_idx, val_idx) in enumerate(fold_splits):
    fold_train = train_val_df.iloc[train_idx]
    fold_val = train_val_df.iloc[val_idx]
    
    print(f"Fold {fold_idx + 1}:")
    print(f"  Train: {len(fold_train):,} samples ({len(fold_train[fold_train['has_sp']]):,} SP, {len(fold_train[~fold_train['has_sp']]):,} NO_SP)")
    print(f"  Val:   {len(fold_val):,} samples ({len(fold_val[fold_val['has_sp']]):,} SP, {len(fold_val[~fold_val['has_sp']]):,} NO_SP)")
```

```python
# Helper function for moderate oversampling (to median, not max)
def oversample_to_median(df, target_col="type", random_state=69):
    """Oversample minority classes to median class count (more moderate than max)."""
    counts = df[target_col].value_counts()
    target_count = int(counts.median())
    
    print(f"Class distribution before: {dict(counts)}")
    print(f"Target count (median): {target_count}")
    
    dfs_to_concat = []
    for label in counts.index:
        subset = df[df[target_col] == label]
        current_count = len(subset)
        
        if current_count < target_count:
            # Oversample minority classes to reach median
            oversampled = subset.sample(n=target_count, replace=True, random_state=random_state)
            dfs_to_concat.append(oversampled)
        else:
            # Keep majority classes as-is (no undersampling)
            dfs_to_concat.append(subset)
    
    result = pd.concat(dfs_to_concat, ignore_index=True)
    result = result.sample(frac=1, random_state=42).reset_index(drop=True)
    
    print(f"Class distribution after: {dict(result[target_col].value_counts())}")
    return result

# Save each fold as separate CSV files with oversampled training data
for fold_idx, (train_idx, val_idx) in enumerate(fold_splits):
    # Get fold data
    fold_train_df = train_val_df.iloc[train_idx].copy()
    fold_val_df = train_val_df.iloc[val_idx].copy()
    
    # Apply moderate oversampling to training set only
    print(f"\n--- Fold {fold_idx + 1} ---")
    fold_train_oversampled = oversample_to_median(fold_train_df)
    
    # Save oversampled train, original val (validation should not be oversampled)
    fold_train_oversampled.to_csv(f"aufgabe3/3-fold/fold_{fold_idx + 1}_train_oversampled.csv", index=False)
    fold_val_df.to_csv(f"aufgabe3/3-fold/fold_{fold_idx + 1}_val_oversampled.csv", index=False)
    
    print(f"Result: {len(fold_train_df):,} → {len(fold_train_oversampled):,} train, {len(fold_val_df):,} val")

print(f"\n✓ All {n_folds} folds saved!")
print(f"Test set saved separately: {len(test_df):,} samples")
```
